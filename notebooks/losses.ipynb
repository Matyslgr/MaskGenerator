```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import sys
import os

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from mask_generator.loss.losses import BoundaryLoss
```

```python
pred = torch.randn(1, 1, 128, 364, device='cuda')
gt = torch.zeros(1, 128, 364, device='cuda')
gt[:, 60:62, :] = 1  # Ligne horizontale simul√©e
```

```python
loss_fn = BoundaryLoss(theta0=3, theta=5)
gt_b, pred_b, gt_b_ext, pred_b_ext = loss_fn(pred, gt.long(), return_maps=True)
```

```python
plt.figure(figsize=(20, 6))
plt.subplot(1, 4, 1)
plt.title("GT Borders")
plt.imshow(gt_b[0, 0].cpu(), cmap='gray')

plt.subplot(1, 4, 2)
plt.title("Pred Borders")
plt.imshow(pred_b[0, 0].cpu(), cmap='gray')

plt.subplot(1, 4, 3)
plt.title("GT Extended")
plt.imshow(gt_b_ext[0, 0].cpu(), cmap='gray')

plt.subplot(1, 4, 4)
plt.title("Pred Extended")
plt.imshow(pred_b_ext[0, 0].cpu(), cmap='gray')

plt.show()
```
