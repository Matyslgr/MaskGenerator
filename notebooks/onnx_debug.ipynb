```python
import onnx
import sys
import os

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
```

```python
home_path = os.path.expanduser("~")
model_path = os.path.join(home_path, 'Téléchargements', 'model.onnx')
if not os.path.exists(model_path):
    print(f"Model file not found at {model_path}. Please check the path.")
model = onnx.load(model_path)
```

```python
onnx.checker.check_model(model)
print("IR version:", model.ir_version)
print("Producer:", model.producer_name)
print("Opset version:", model.opset_import[0].version)
print("Nombre de noeuds:", len(model.graph.node))
```

```python
for i, node in enumerate(model.graph.node):
    print(f"Node {i}:")
    print(f"  OpType: {node.op_type}")
    print(f"  Name: {node.name}")
    print(f"  Inputs: {node.input}")
    print(f"  Outputs: {node.output}")
    print("-" * 30)
```

```python
for initializer in model.graph.initializer:
    print(f"Initializer: {initializer.name} - shape: {initializer.dims} - dtype: {initializer.data_type}")
```
