```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

IMAGE_PATH = "/root/MaskGenerator/Datasets/CARLANE/MoLane/data/val/target/straighttrack/gray/dark/0035_label.png"

# Load the image
mask = cv2.imread(IMAGE_PATH)


# Convert BGR to RGB
mask = cv2.cvtColor(mask, cv2.IMREAD_GRAYSCALE)
mask = (mask >= 1).astype(np.uint8)
mask = mask * 255  # Convert to binary mask (0 or 255)

plt.figure(figsize=(10, 10))
plt.imshow(mask, cmap='gray')
plt.axis("off")
plt.show()
```
